```python
def cumulative_runs(match):    
    cumulative_runs = []
    curr_runs = 0
    for i in range(match.shape[0]):
        curr_runs += match['total_runs'][i]
        cumulative_runs.append(curr_runs)
    return np.array(cumulative_runs)
```

```python
matches = df['match_id'].unique()
matches.size
```

```python
match_dataframes = []
for match in matches:
    match_info = df.loc[df['match_id'] == match].reset_index()
    wickets = 0
    runs_last_7 = []
    runs_past_7overs = []
    backWeek = 0
    for i, ball in enumerate(match_info['dismissal_kind']):
        if ball != ball:
            pass
        else:
            wickets += 1
        match_info['over'][i] = match_info['over'][i] + 0.1*match_info['ball'][i]
        match_info['dismissal_kind'][i] = wickets
        runs_past_7overs.append(match_info['total_runs'][i])
        if len(runs_past_7overs) > 7*6:
            runs_past_7overs.pop(0)
        runs_last_7.append(sum(runs_past_7overs))
        
    c = cumulative_runs(match_info)
    match_info = match_info.assign(Total=pd.Series(np.ones(len(match_info['match_id']))*c[-1]).values)
    match_info = match_info.assign(runs_last_7=pd.Series(np.array(runs_last_7)).values)
    match_dataframes.append(match_info)
df = pd.concat(match_dataframes).reset_index()
```

# PolyFit - Only extrapolation

```python
def prediction(match):
    match_info = df.loc[df['match_id'] == match].reset_index()
    till_over = 8
    match_train = match_info.loc[match_info['over'] < till_over].reset_index()

    c = cumulative_runs(match_info)
    c_train = cumulative_runs(match_train)
    x = np.linspace(0, match_info.shape[0]-1, num=match_info.shape[0])
    x_train = np.linspace(0, match_train.shape[0]-1, num=match_train.shape[0])

    fit = np.polyfit(x_train, c_train, 1)
    line = np.poly1d(fit)
    pred = line(x)
    return pred, c, c_train, x, x_train
```

```python
def plot_pred(match):    
    pred, c, c_train, x, x_train = prediction(match)
    print("Prediction:", pred[-1], ", Actual:", c[-1])
    fig, axis = plt.subplots()
    axis.plot(x, c)
    axis.plot(x_train, c_train)
    axis.plot(x, pred)
    plt.show()
```

```python
totalLoss = 0
maxLoss = -1
maxLossIdx = -1
for match in matches:
    pred, c, _, _, _ = prediction(match)
    loss = (pred[-1]-c[-1])**2
    print("Match:", match, "Prediction:", pred[-1], ", Actual:", c[-1], ", Loss:", loss)
    if maxLoss < loss:
        maxLoss = loss
        maxLossIdx = match
    totalLoss += loss
print(totalLoss/len(matches))
print("Max Deviation in match", maxLossIdx)
```

```python
encoded_df = pd.get_dummies(data=df, columns=['batting_team', 'bowling_team'])
encoded_df.columns
```

```python
encoded_df
```

```python
X_train = encoded_df.drop(labels=['Total', 'total_runs', 'over'], axis=1)[encoded_df['match_id']<500]
X_test = encoded_df.drop(labels=['Total', 'total_runs', 'over'], axis=1)[encoded_df['match_id']>=500]

y_train = encoded_df[encoded_df['match_id']<500]['Total'].values
y_test = encoded_df[encoded_df['match_id']>=500]['Total'].values

X_train.drop(labels='match_id', axis=True, inplace=True)
X_test.drop(labels='match_id', axis=True, inplace=True)

print("Training set: {} and Test set: {}".format(X_train.shape, X_test.shape))
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
linear_regressor = LinearRegression()
linear_regressor.fit(X_train,y_train)
```

```python
y_pred_train = linear_regressor.predict(X_train)
y_pred_lr = linear_regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse, accuracy_score
print("Mean Absolute Error (MAE): {}".format(mae(y_train, y_pred_train)))
print("Mean Squared Error (MSE): {}".format(mse(y_train, y_pred_train)))
print("Root Mean Squared Error (RMSE): {}".format(np.sqrt(mse(y_train, y_pred_train))))
print("---- Linear Regression - Model Evaluation ----")
print("Mean Absolute Error (MAE): {}".format(mae(y_test, y_pred_lr)))
print("Mean Squared Error (MSE): {}".format(mse(y_test, y_pred_lr)))
print("Root Mean Squared Error (RMSE): {}".format(np.sqrt(mse(y_test, y_pred_lr))))
```
